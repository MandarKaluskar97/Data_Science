```python
# Columns -> 'StandardHours', 'EmployeeCount', 'Over18' and 'EmployeeNumber' can be dropped
df.drop(['StandardHours'], axis=1, inplace=True)
df.drop(['EmployeeCount'], axis=1, inplace=True)
df.drop(['Over18'], axis=1, inplace=True)
df.drop(['EmployeeNumber'], axis=1, inplace=True)
df.shape
```

### Duplicates check

```python
df.drop_duplicates(inplace=True)
print(df.shape)
df.index
```

### Correlation Check

```python
attrition_count = pd.DataFrame(df['Attrition'].value_counts())
attrition_count
```

```python
plt.pie(attrition_count['Attrition'] , labels = ['No' , 'Yes'] , explode = (0.2,0), autopct='%1.1f%%')
# Looks like the dataset is disproportionate with close to 84% of the records having "No" Attrition
```

```python
df_new = df.select_dtypes(include=np.number)
print(df_new.shape)
df_new.columns
```

```python
corrmat = df_new.corr()
corrmat
```

```python
# The dataset contains large number of columns to be accomodated in a single HeatMap, hence dividing it in 3 HeatMaps

h1 = corrmat.loc['Age':'NumCompaniesWorked', 'Age':'NumCompaniesWorked']
h2 = corrmat.loc['PercentSalaryHike':, 'Age':'NumCompaniesWorked']
h3 = corrmat.loc['PercentSalaryHike':, 'PercentSalaryHike':]

sns.set(rc = {'figure.figsize':(15,8)})
sns.heatmap(h1, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 11}, cmap="YlGnBu", 
            linewidths=0.5, linecolor='blue')

# MonthlyIncome and JobLevel are highly correlated
```

```python
sns.heatmap(h2, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 11}, cmap="YlGnBu", 
            linewidths=0.5, linecolor='blue')

# TotalWorkingYears and JobLevel are highly correlated
# TotalWorkingYears and MonthlyIncome are highly correlated
```

```python
sns.heatmap(h3, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 11}, cmap="YlGnBu", 
            linewidths=0.5, linecolor='blue')

# PercentSalaryHike and PerformanceRating are highly correlated
# YearsInCurrentRole and YearsAtCompany are highly correlated
# YearsWithCurrManager and YearsAtCompany are highly correlated
```

<u>Correlation Outcome</u>:

- 'MonthlyIncome' and 'JobLevel' are highly correlated => r = 0.95
- 'TotalWorkingYears' and 'JobLevel' are highly correlated => r = 0.78
- 'TotalWorkingYears' and 'MonthlyIncome' are highly correlated => r = 0.77
- 'PercentSalaryHike' and ''PerformanceRating' are highly correlated => r = 0.77
- 'YearsInCurrentRole' and 'YearsAtCompany' are highly correlated => r = 0.76
- 'YearsWithCurrManager' and 'YearsAtCompany' are highly correlated => r = 0.77

```python
# Dropping the correlated columns:
df_final = df.drop(['JobLevel','TotalWorkingYears','YearsInCurrentRole', 'YearsWithCurrManager' , 'PercentSalaryHike'], axis=1)
df_final # It is final cleaned dataset which we will be working on
```

# Finding Insights from data

```python
df_final['Attrition'] = df_final.Attrition.apply(lambda x: 1 if x == 'Yes' else 0)
df_final
```

```python
ax = sns.histplot(data = df_final, x = "Age", hue = "Attrition", palette = "muted")
ax.set_title("Age of Employees", pad = 45)
ax.set_ylabel(None)
ax.set_xlabel("Age in Years", labelpad = 25)
plt.text(x = 45, y = 110, s =  "Average age is 37 years", fontweight = 'semibold')

# Majority of the employees leaving are less than 40 years old 
```

```python
plt.figure(figsize=(20,30))
plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.4, 
                    hspace=0.4)

plt.subplot(3,1,1)
ax = sns.countplot(x = 'BusinessTravel', hue = 'Attrition', data = df_final)
ax.set_title('Attrition rate based on Business Travel', fontsize = 25)
ax.set_xlabel('Business Travel', fontsize=20)
ax.set_ylabel('Count', fontsize=20)
for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy() 
    ax.annotate(f'{height}', (x + width/2, y + height*1.02), ha='center')
    

plt.subplot(3,1,2)
ax = sns.countplot(x = 'Department', hue = 'Attrition', data = df_final)
ax.set_title('Department specific Attrition rate', fontsize = 25)
ax.set_xlabel('Department', fontsize=20)
ax.set_ylabel('Count', fontsize=20)
for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy() 
    ax.annotate(f'{height}', (x + width/2, y + height*1.02), ha='center')
    

plt.subplot(3,1,3)
ax = sns.countplot(x = 'Gender', hue = 'Attrition', data = df_final)
ax.set_title('Gender-specific Attrition rate', fontsize = 25)
ax.set_xlabel('Gender', fontsize=20)
ax.set_ylabel('Count', fontsize=20)
for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy() 
    ax.annotate(f'{height}', (x + width/2, y + height*1.02), ha='center')
    
    
# Majority of Employees travel rarely for business purpose.
# The Research and Development Department has the highest number of Employees, most of them having Life Sciences and Medical Degrees.
```

```python
names = ["JobInvolvement", "JobSatisfaction", "PerformanceRating"]
for name in names:
    df_final[name].replace({1 : "Low", 2 : "Medium", 3 : "High", 4 : "Very High"}, inplace = True)
```

```python
ax = sns.countplot(x = 'JobSatisfaction', hue = 'Attrition', data = df_final, order = ['Low', 'Medium', 'High', 'Very High'])
ax.set_title('Job Satisfaction by Attrition', fontsize = 25)
for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy() 
    ax.annotate(f'{height}', (x + width/2, y + height*1.02), ha='center')

# No clear pattern followed here, the employees with "High" Job Satisfaction have contributed most towards the Attrition, 
# followed by the employees with "Low" Job Satisfaction.  
```

```python
ax = sns.countplot(x = 'JobInvolvement', hue = 'Attrition', data = df_final, order = ['Low', 'Medium', 'High', 'Very High'])
ax.set_title('Job Involvement by Attrition', fontsize = 25)
for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy() 
    ax.annotate(f'{height}', (x + width/2, y + height*1.02), ha='center')

# Employees with "High" Job Involvement contribute the most to Attrition
```

```python
ax = sns.countplot(x = 'PerformanceRating', hue = 'Attrition', data = df_final, order = ['Low', 'Medium', 'High', 'Very High'])
ax.set_title('Performance Rating by Attrition', fontsize = 25)
for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy() 
    ax.annotate(f'{height}', (x + width/2, y + height*1.02), ha='center')

# Clearly, almost 85% of the employees who left the company had a Performance Rating of '3'
```
